#Analysis of Traffic Accident Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
```

```python
data=pd.read_csv("/content/accidents_2005_to_2007.csv")
data.head(10)
```

```python
# Select relevant columns for analysis
selected_columns = ['Weather_Conditions', 'Road_Surface_Conditions', 'Special_Conditions_at_Site', 'Urban_or_Rural_Area', 'Time']
selected_data = data[selected_columns].copy()
```

```python
# Exploratory Data Analysis (EDA)

# Visualize Weather Conditions with Colors
plt.figure(figsize=(12, 8))

# Choose a color palette (you can customize the colors if needed)
color_palette = sns.color_palette("husl", n_colors=len(selected_data['Weather_Conditions'].unique()))

sns.countplot(x='Weather_Conditions', data=selected_data, palette=color_palette)
plt.title('Accident Counts by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Accident Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize Road Surface Conditions
plt.figure(figsize=(8, 6))
sns.countplot(x='Road_Surface_Conditions', data=selected_data)
plt.title('Accident Counts by Road Surface Conditions')
plt.xlabel('Road Surface Conditions')
plt.ylabel('Accident Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Convert 'Time' column to datetime format
data['Time'] = pd.to_datetime(data['Time'])

# Extract hour information from the 'Time' column
data['Hour'] = data['Time'].dt.hour

# Plotting accident counts by hour
plt.figure(figsize=(10, 6))
sns.countplot(x='Hour', data=data)
plt.title('Accident Counts by Time of Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Accident Count')
plt.show()
```

```python
# Visualize Special Conditions at Site
plt.figure(figsize=(8, 6))
sns.countplot(x='Special_Conditions_at_Site', data=selected_data)
plt.title('Accident Counts by Special Conditions at Site')
plt.xlabel('Special Conditions at Site')
plt.ylabel('Accident Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize Urban or Rural Area
plt.figure(figsize=(6, 4))
sns.countplot(x='Urban_or_Rural_Area', data=selected_data)
plt.title('Accident Counts by Urban or Rural Area')
plt.xlabel('Urban or Rural Area')
plt.ylabel('Accident Count')
plt.show()
```
